```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

file_path = "RTA Dataset.csv"   # Make sure the file is in the same folder
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.hour
df[['Time', 'Hour']].head()
```

```python
hour_counts = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(10,5))
hour_counts.plot(kind='bar')
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
weather_counts = df['Weather_conditions'].value_counts()

plt.figure(figsize=(10,5))
weather_counts.plot(kind='bar')
plt.title("Accidents by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
road_counts = df['Road_surface_conditions'].value_counts()

plt.figure(figsize=(10,5))
road_counts.plot(kind='bar')
plt.title("Accidents by Road Surface Condition")
plt.xlabel("Road Surface Condition")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
area_counts = df['Area_accident_occured'].value_counts().head(10)

plt.figure(figsize=(10,5))
area_counts.plot(kind='bar')
plt.title("Top 10 Accident Hotspots by Area")
plt.xlabel("Area")
plt.ylabel("Number of Accidents")
plt.show()
```
